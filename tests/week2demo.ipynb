```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/Winter_medals2022-11-03.csv")
df
```

```python
relevent_df = df[["Country", "Gold", "Silver", "Bronze"]] 
relevent_df
```

```python
grouped_df = relevent_df.groupby(["Country"]).sum()
grouped_df.head()
```

```python
total = grouped_df.sum(axis=1)
grouped_df["Total"] = total
grouped_df.head()
```

```python
sorted_df = grouped_df.sort_values("Total", ascending = False)
sorted_df.head()
```

```python
df \
    [["Country", "Gold"]] \
    .groupby("Country") \
    .sum() \
    .sort_values("Gold") \
    .plot.bar()
```
