```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### What are the most prevelant species?

```python
most_common_spicies = df['Artnamn'].value_counts().head(5)
most_common_spicies
```

 ### What is the monthly distribution of the top 3 most prevelant species?

```python
# Step 1: Identify the top 3 most prevalent species
top_3_species = df['Artnamn'].value_counts().head(3)
top_3_names = top_3_species.index  # Get the species names

# Step 2: Extract the month from 'Startdatum'
df['Startdatum'] = pd.to_datetime(df['Startdatum'], format='%Y-%m-%d')  # Convert to datetime
df['Month'] = df['Startdatum'].dt.month  # Extract the month as an integer

# Step 3: Filter data for only the top 3 species
filtered_df = df[df['Artnamn'].isin(top_3_names)]

# Step 4: Group by species and month, then count observations
monthly_distribution = (
    filtered_df.groupby(['Artnamn', 'Month'])
    .size()
    .reset_index(name='Count')  # Convert counts to a new column
)

# Step 5: Plot the monthly distribution
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=monthly_distribution,
    x='Month',
    y='Count',
    hue='Artnamn',
    marker='o',
    palette='Set2'
)

# Annotate the top 3 species in the title dynamically
title = f"Monthly Distribution of the Top 3 Most Prevalent Species:\n{', '.join(top_3_names)}"
plt.title(title, fontsize=14)
plt.xlabel('Month')
plt.ylabel('Number of Observations')
plt.xticks(range(0, 13))  
plt.legend(title='Species')
plt.grid()
plt.tight_layout()
plt.show()
```

### What spicies have the most spottings alone, in groups of 2, groups of 3, ...,groups of 10?

```python
# Step 1: Ensure 'Antal' is numeric (in case it's stored as a string)
df['Antal'] = pd.to_numeric(df['Antal'], errors='coerce')

# Step 2: Group by 'Antal' and find the most common species in each group
most_common_species_per_antal = (
    df.groupby('Antal')['Artnamn']
    .apply(lambda group: group.value_counts().idxmax())
    .reset_index(name='Most_Common_Species')
)

# Step 3: Add the count of the most common species for each 'Antal'
most_common_counts = (
    df.groupby(['Antal', 'Artnamn']).size()
    .reset_index(name='Count')
)

# Merge the counts with the most common species
plot_data = pd.merge(
    most_common_species_per_antal,
    most_common_counts,
    left_on=['Antal', 'Most_Common_Species'],
    right_on=['Antal', 'Artnamn']
)

# Step 4: Filter for 'Antal' less than or equal to 10
plot_data = plot_data[plot_data['Antal'] <= 10]

# Step 5: Create the plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=plot_data,
    x='Antal',
    y='Count',
    hue='Most_Common_Species',
    dodge=False,
    palette='Set2'
)

plt.title("Most Common Species by Quantity", fontsize=16)
plt.xlabel("Quantity", fontsize=12)
plt.ylabel("Number of Observations", fontsize=12)
plt.legend(title="Species", loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### For the top 10 most common spicies, in what quantity are they most usually spotted?

```python
# Step 1: Ensure 'Antal' is numeric (if necessary)
df['Antal'] = pd.to_numeric(df['Antal'], errors='coerce')

# Step 2: Find the top 10 most common species
top_10_species = df['Artnamn'].value_counts().head(10).index

# Step 3: Filter the DataFrame to include only the top 10 species
filtered_df = df[df['Artnamn'].isin(top_10_species)]

# Step 4: Find the most common quantity ('Antal') for each species
most_common_quantity = (
    filtered_df.groupby('Artnamn')['Antal']
    .apply(lambda group: group.value_counts().idxmax())
    .reset_index(name='Most_Common_Quantity')
)

# Step 5: Count occurrences of the most common quantity for each species
quantity_counts = (
    filtered_df.groupby(['Artnamn', 'Antal'])
    .size()
    .reset_index(name='Count')
)

# Merge to keep only the most common quantity for each species
plot_data = pd.merge(
    most_common_quantity,
    quantity_counts,
    left_on=['Artnamn', 'Most_Common_Quantity'],
    right_on=['Artnamn', 'Antal']
)

# Step 6: Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=plot_data,
    x='Artnamn',
    y='Most_Common_Quantity',
    palette='Set3'
)

plt.title("Most Common Quantity by Top 10 Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Most Common Quantity ('Antal')", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
